```python
class Student:
    total = 0

    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
        Student.total += 1


n = int(input("Enter number of Students"))

p = []
for i in range(n):
    name = input("Enter student's name: ")
    age = int(input("Enter student's age: "))
    grade = float(input("Enter student's grade: "))
    p1 = Student(name, age, grade)
    p.append(p1)

total_grade = 0
for i in range(n):
    total_grade += p[i].grade

av = total_grade / Student.total

if av < 0 or av > 10:
    print("CO sinh vien nao do diem sai")
else:
    if av > 9:
        print("sx")
    elif av > 8:
        print("gioi")
    elif av > 7:
        print("Kha")
    elif av > 5:
        print("Pass")
    else:
        print("fail")


print(av)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def eat(self):
        print("eat out")


# student
class Student(Person):
    def __init__(self, name, age, id):
        super().__init__(name, age)
        self.__id = id

    def get_id(self):
        return self.__id


s1 = Student("Tung", 24, "AUH14948")
print(s1.eat())
```

Người dùng nhập vào từng sinh viên một với tên, tuổi, điểm số. In ra màn hình xem có bao nhiêu sinh viên, và điểm trung bình của các sinh viên đó.


```python
class Student:
    total = 0

    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
        Student.total += 1


s = []

while True:
    name = input("Enter student's name")
    age = int(input("Enter student's age"))
    grade = float(input("Enter student's grade"))
    s1 = Student(name, age, grade)
    s.append(s1)
    check = input("Do you want to enter more? y/n")
    if check == "n":
        break

total_grade = 0

for i in range(len(s)):
    total_grade += s[i].grade

av = total_grade / Student.total
print(av)
print(total_grade)
```

```python
class Student:
    def __init__(self, name, age, id):
        self.name = name
        self.age = age
        self.__id = id #private

    def get_id(self):
        return self.__id
    
    def set_id(self, new_id):
        self.__id = new_id

s1 = Student("Tuan", 18, "AUH2828")
print(s1.get_id())
s1.set_id("AUH1234")
print(s1.get_id())
```

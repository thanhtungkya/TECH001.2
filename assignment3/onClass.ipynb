```python
n = 6

#fibo(n) = fibo(n - 1) + fibo(n - 2)
def fibo(n:int)->int:
    if(n == 1 or n == 2):
        return 1
    else:
        return fibo(n - 1) + fibo(n - 2)

print(fibo(6))
```
